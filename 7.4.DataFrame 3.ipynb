## .apply Method

```python
def con(number):
    return str(number) + ' MILLIONS!'

bond['Box_Office'] = bond['Box_Office'].apply(con)
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return 'The best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'I have no clue'
```

```python
bond.apply(good_movie, axis=1)
```
