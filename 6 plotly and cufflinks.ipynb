```python
df = pd.DataFrame(np.random.randn(100,3), columns=['A','B','C'])
df.head()
```

# Iplot

```python
df.iplot()
```

```python
df['A'] = df['A'].cumsum()+20
df['B'] = df['B'].cumsum()+20
df['C'] = df['C'].cumsum()+20
```

```python
df.iplot()
```

```python
plt.plot(df)
```

```python
df.plot()
```

```python
df.iplot(x='A' , y='B')
```

# Scatter Plot

# 3D Surface

```python
df3 = pd.DataFrame({'X':[10,20,30,10,20] , 'Y':[10,20,30,10,20] , 'Z':[10,20,30,10,20]})
df3
```

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface' , colorscale='rdylbu')
```

```python
cf.datagen.sinwave(10,0.25).iplot(kind='surface')
```

# Scatter3D

```python
cf.datagen.scatter3d(2,150 , mode='stocks').iplot(kind='scatter3d' , x='x', y='y' , z='z')
```

# Spread

```python
df[['A', 'B']].iplot(kind='spread') #How it changes 1 vs other 
```

# Histogram

```python
df.iplot(kind='hist' , bins=25 , barmode='group' , bargap=0.5)
```

```python
df.iplot(kind='hist' , bins=25 , barmode='stack' , bargap=0.5)
```

```python
df.iplot(kind='hist' , bins=25 , barmode='overlay' , bargap=0.5)
```

# Bubble

```python
cf.datagen.bubble3d(5,4, mode='stocks').iplot(kind='bubble3d' , x='x' , y='y' , z='z' , size='size')
```

# heatmap

```python
cf.datagen.heatmap(20,20).iplot(kind='heatmap' , colorscale='spectral' , title='Cufflinks - heatmap')
```
