## Bar Charts

```python
x = data.DataReader(name='GOOG', data_source='yahoo', start='2007-07-01', end='2008-12-03')
x.head()
```

```python
def rank(stock_price):
    if stock_price <= 200:
        return 'Poor'
    elif stock_price > 200 and stock_price <= 500:
        return 'Satisfactory'
    else:
        return 'Steller'
```

```python
x['Close'].apply(rank)
```

```python
x['Close'].apply(rank).value_counts()
```

```python
x['Close'].apply(rank).value_counts().plot(kind='bar')
```

```python
x['Close'].apply(rank).value_counts().plot(kind='barh')
```

## Pie Charts

```python
apple = data.DataReader(name='AAPL', data_source='yahoo', start='2012-01-01', end='2016-01-01')
```

```python
apple.head()
```

```python
apple['Close'].mean()
```

```python
def rank_apple(stock_price):
    if stock_price >= 90.54555943756407:
        return 'Above Average'
    else:
        return 'Below Average'
```

```python
plt.style.use('ggplot')
apple['Close'].apply(rank_apple)
```

```python
apple['Close'].apply(rank_apple).value_counts()
```

```python
apple['Close'].apply(rank_apple).value_counts().plot(kind='pie', legend=True)
```

## Histogram

```python
apple['Close'].apply(rank_apple).value_counts().plot(kind='hist', legend=True)
```

```python
apple['Close'].apply(rank_apple).value_counts().plot(kind='hist', bins=3)
```
