```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot( x='timepoint' , y='signal' , kind='line' , data=fmri , ci=False) #Confidence interval - light blue region ci=True
```

```python
sns.relplot( x='timepoint' , y='signal' , kind='line' , data=fmri , ci='sd')  #Confidence interval = standard deviation
```

```python
sns.relplot( x='timepoint' , y='signal' , estimator=None , kind='line' , data=fmri , ci=True) #when estimator = None -> It will plot real data set
```

```python
sns.relplot( x='timepoint' , y='signal' , hue='event' , kind='line' , data=fmri )
```

```python
sns.relplot( x='timepoint' , y='signal' , hue='region' , style='event' , kind='line' , data= fmri)
```

```python
fmri['region'].value_counts()
```

```python
fmri['event'].value_counts()
```

# Violin Plot

```python
#Violin plot = KDE plot + boxplot
```

```python
sns.catplot( x='total_bill' , y='day' , hue='time' , kind='violin' , data=tips )
```
