# Controlling Ploted figure aesthtics 
- figure styling
- axes styling
- color palettes

```python
def sinplot(flip = 1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
sns.set_style('whitegrid')
sinplot()
```

```python
sns.set_style('white' and 'ticks')
sinplot()
```

```python
sns.set_style('white' and 'ticks' , {'axes.grid':True ,  'xtick.direction': 'in'})
sinplot()
sns.despine() # remove rght and top
sns.despine(left=True)
```
