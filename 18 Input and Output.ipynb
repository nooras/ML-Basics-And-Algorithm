## Quick Object Conversions

```python
URL = 'https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv'
babyName = pd.read_csv(URL)
babyName.head()
```

```python
babyName['Child\'s First Name'].tolist()
```

```python
babyName['Child\'s First Name'].to_frame()
```

```python
', '.join(str(name) for name in babyName['Child\'s First Name'])
```

```python
babyName.to_csv('pandas/BabyName.csv')
```

```python
babyName.to_csv('pandas/BabyName2.csv', index=False, columns=['Gender'], encoding='utf-8')
```

## Export excel 

```python
girl = babyName[babyName['Gender'] == 'FEMALE']
boy = babyName[babyName['Gender'] == 'MALE']
```

```python
excel_file = pd.ExcelWriter('pandas/BabName.xlsx')
```

```python
girl.to_excel(excel_file, sheet_name='Girls', index=False)
boy.to_excel(excel_file, sheet_name='Boys', index=False, columns=['Gender'])
```

```python
excel_file.save()
```
