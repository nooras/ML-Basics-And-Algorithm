# Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Statistical analysis is a process of understanding how variable in the dataset related to eachother and how does 
# it relate sensibly prints on the other variable
```

```python
'''
1.Real Plot
2.Relational plot
3.Scattered plot
4.Line plot
5.Categorical plot
'''
```

```python
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')   #In the form of pandas datframe
tips.tail()
```

```python
#Relational Plot
sns.relplot(x = 'total_bill',y = 'tip',data = tips,hue = 'smoker',style = 'time') #Categorical paletes
```

```python
dir(sns.FacetGrid) #We can apply these things
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,hue = 'size',style = 'time') #qualititive palletees
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,hue = 'size',palette='ch:r=0.8,l=0.95')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,size = 'size')
```

```python
sns.relplot(x = 'total_bill',y = 'tip',data = tips,size = 'size',sizes = (15,200))
```
